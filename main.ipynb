#### Média Salarial Mensal por Grupo de Idade e Região

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path_age = './média-salarial-rendimento-por-estado-por-grupo-de-idade.xlsx'

# Limpeza dos dados

# Importando o csv e ignoramos as 3 primeiras linhas que são cabeçalhos
df_age = pd.read_excel(file_path_age, skiprows=3) 

# Renomeando as colunas
df_age.columns = ["Região", "14 a 17 anos", "18 a 24 anos", "25 a 39 anos", "40 a 59 anos", "60 ou mais"] 

# Removendo as 2 primeiras linhas e a última dos dados que são o total que nos interessa e uma linha mesclada.
df_age = df_age.iloc[2:]
df_age = df_age.iloc[:-1]

# Definindo a coluna "Região" como o indice da tabela
df_age.set_index("Região", inplace=True)

# Transpondo a tabela para que as regiões sejam colunas para facilitar a visualização e plotagem
df_age = df_age.T

# Configura plotação
df_age.plot(kind='bar', figsize=(10, 6), width=0.8)

# Estiliza a plotação
plt.title('Média Salarial por Faixa Etária e Região', fontsize=14)
plt.xlabel('Faixa Etária', fontsize=12)
plt.ylabel('Média Salarial Mensal (R$)', fontsize=12)
plt.xticks(rotation=0)

plt.legend(title="Região", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()


```

#### Média Salarial Mensal por Genêro e Região

#### Média Salarial Mensal por Nível de Escolaridade

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path_escolaridade = './média-salarial-por-estado-por-nivel-escolaridade.xlsx'

# Limpeza dos dados

# Importando o csv e ignoramos as 3 primeiras linhas que são cabeçalhos
df_escolaridade = pd.read_excel(file_path_escolaridade, skiprows=3) 

# Renomeando as colunas
df_escolaridade.columns = [
    "Região", 
    "Sem instrução e menos de 1 ano de estudo", 
    "Ensino fundamental incompleto ou equivalente", 
    "Ensino fundamental completo ou equivalente", 
    "Ensino médio incompleto ou equivalente", 
    "Ensino médio completo ou equivalente",
    "Ensino superior incompleto ou equivalente",
    "Ensino superior completo ou equivalente",
    "Não determinado",
] 

# Definindo a coluna "Região" como o indice da tabela
df_escolaridade.set_index("Região", inplace=True)

# Deixando apenas a coluna da média do Brasil
df_escolaridade = df_escolaridade.iloc[1:]
df_escolaridade = df_escolaridade.iloc[:1]
df_escolaridade.drop(columns=['Não determinado'], inplace=True)


# Configura plotação
df_escolaridade.plot(kind='bar', figsize=(10, 6), width=0.8)

# Estiliza a plotação
plt.title('Média Salarial por Nível de Escolaridade no Brasil', fontsize=14)
plt.xlabel('Nível de Escolaridade', fontsize=12)
plt.ylabel('Média Salarial Mensal (R$)', fontsize=12)
plt.xticks(rotation=0)

plt.legend(title="Região", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()


```
