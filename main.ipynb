```python
import pandas as pd
import matplotlib.pyplot as plt

file_path_age = './mdia-salarial-rendimento-por-estado-por-grupo-de-idade.xlsx'

# Limpeza dos dados

# Importando o csv e ignoramos as 3 primeiras linhas que so cabealhos
df_age = pd.read_excel(file_path_age, skiprows=3) 

# Renomeando as colunas
df_age.columns = ["Regio", "14 a 17 anos", "18 a 24 anos", "25 a 39 anos", "40 a 59 anos", "60 ou mais"] 

# Removendo as 2 primeiras linhas e a ltima dos dados que so o total que nos interessa e uma linha mesclada.
df_age = df_age.iloc[2:]
df_age = df_age.iloc[:-1]

# Definindo a coluna "Regio" como o indice da tabela
df_age.set_index("Regio", inplace=True)

# Transpondo a tabela para que as regies sejam colunas para facilitar a plotagem
df_age = df_age.T

# Configura plotao
df_age.plot(kind='bar', figsize=(10, 6), width=0.8)

# Estiliza a plotao
plt.title('Mdia Salarial por Faixa Etria e Regio', fontsize=14)
plt.xlabel('Faixa Etria', fontsize=12)
plt.ylabel('Mdia Salarial Mensal (R$)', fontsize=12)
plt.xticks(rotation=0)

plt.legend(title="Regio", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path_escolaridade = './mdia-salarial-por-estado-por-nivel-escolaridade.xlsx'

# Limpeza dos dados

# Importando o csv e ignoramos as 3 primeiras linhas que so cabealhos
df_escolaridade = pd.read_excel(file_path_escolaridade, skiprows=3) 

# Renomeando as colunas
df_escolaridade.columns = [
    "Regio", 
    "Sem instruo e menos de 1 ano de estudo", 
    "Ensino fundamental incompleto ou equivalente", 
    "Ensino fundamental completo ou equivalente", 
    "Ensino mdio incompleto ou equivalente", 
    "Ensino mdio completo ou equivalente",
    "Ensino superior incompleto ou equivalente",
    "Ensino superior completo ou equivalente",
    "No determinado",
] 

# Definindo a coluna "Regio" como o indice da tabela
df_escolaridade.set_index("Regio", inplace=True)

# Deixando apenas a coluna da mdia do Brasil
df_escolaridade = df_escolaridade.iloc[1:]
df_escolaridade = df_escolaridade.iloc[:1]
df_escolaridade.drop(columns=['No determinado'], inplace=True)


# Configura plotao
df_escolaridade.plot(kind='bar', figsize=(10, 6), width=0.8)

# Estiliza a plotao
plt.title('Mdia Salarial por Nvel de Escolaridade no Brasil', fontsize=14)
plt.xlabel('Nvel de Escolaridade', fontsize=12)
plt.ylabel('Mdia Salarial Mensal (R$)', fontsize=12)
plt.xticks(rotation=0)

plt.legend(title="Regio", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()


```
